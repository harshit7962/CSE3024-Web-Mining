```python
#Importing Libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Initiating and designing graph
G = nx.DiGraph()
G.add_edge('1','2')
G.add_edge('1','3')
G.add_edge('1','4')
G.add_edge('2','4')
G.add_edge('2','5')
G.add_edge('3','6')
G.add_edge('4','3')
G.add_edge('4','6')
G.add_edge('4','7')
G.add_edge('5','4')
G.add_edge('5','7')
G.add_edge('7','6')
nx.draw(G, with_labels=True)
plt.show()
```

```python
#Degree Calculation
n_nodes = 7
print("Node\tDegree")
for node in G.nodes():
    print(node, "\t", len(G.in_edges(node))/(n_nodes-1))
```

```python
#Proximity Calculation
distance = []

temp_dis = 0
n = 0
for dest in G.nodes:
    temp_dis = 0
    n = 0
    for src in G.nodes:
        if (nx.has_path(G,src,dest) == True):
            temp_dis = temp_dis + nx.shortest_path_length(G,source = src,target = dest)
            n = n + 1
    if temp_dis == 0:
        distance.append([dest, 0])
    else:
        distance.append([dest, temp_dis/(n - 1)])
        
print("Node\tProximity")
for i in distance:
    print(str(i[0]) + " \t " + str(i[1]))
```
