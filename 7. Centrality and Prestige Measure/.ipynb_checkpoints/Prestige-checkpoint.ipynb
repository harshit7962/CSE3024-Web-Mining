```python
#Importing Libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Initiating and designing graph
G = nx.DiGraph()
G.add_edge('1','2')
G.add_edge('1','3')
G.add_edge('1','4')
G.add_edge('2','4')
G.add_edge('2','5')
G.add_edge('3','6')
G.add_edge('4','3')
G.add_edge('4','6')
G.add_edge('4','7')
G.add_edge('5','4')
G.add_edge('5','7')
G.add_edge('7','6')
nx.draw(G, with_labels=True)
plt.show()
```

```python
n_nodes = 7
print("Node\tDegree")
for node in G.nodes():
    print(node, "\t", len(G.in_edges(node))/(n_nodes-1))
```
